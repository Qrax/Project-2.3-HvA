```python
#Alle benodigde libraries importen.
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import csv
```

```python
#Onze data van het experiment importeren, naar een NumPy array zetten en printen.
data = pd.read_excel(r'Workshop1Excel.xlsx')
data = np.array(data)
print(data)
```

```python
#De voltage en tijds data uit de NumPy array halen.
volt = data[: ,4]
print(volt)

tijd = data[:, 3]
print(tijd)
```

```python
#Plot maken van de data.
plt.figure(figsize=(10, 5), dpi=400)
plt.title("Complete meting van voltage over tijd")
plt.xlabel("Tijd (s)")
plt.ylabel("Voltage (mV)")
plt.plot(tijd, volt)
```

```python
#Tijdsconstante uitrekenen

tau = []
dV = 0
recent = 0

for x in range(len(volt)):
    if (volt[x] > volt[x-1]+0.3 and volt[x-1] != 0):
        top = tijd[x]
        print(x)
        dV = ((volt[x]/100)*37)
        #print(dV)
    if (volt[x] > dV*0.98 and volt[x] < dV*1.02 and x-recent > 10) :
        tau.append(tijd[x]-top)
        recent = x
```

```python
#Alle tijdsconstantes printen.
print(tau)

#Het gemiddelde van alle tijdsconstantes printen
print("Tijdsconstante = " + str(np.mean(tau)) + " seconden")
```
